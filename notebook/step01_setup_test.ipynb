```python
# ============================================
# STEP 01: Setup & Environment Test
# ============================================

import os
from dotenv import load_dotenv


load_dotenv()


hf_token = os.getenv("HUGGINGFACEHUB_API_TOKEN")
pinecone_key = os.getenv("PINECONE_API_KEY")

print("=" * 40)
print("       ENVIRONMENT CHECK")
print("=" * 40)
print(f"HuggingFace Token : {'Loaded' if hf_token else '❌ Missing'}")
print(f"Pinecone API Key  : {'Loaded' if pinecone_key else '❌ Missing'}")
print("=" * 40)
```

```python
from huggingface_hub import InferenceClient

def test_huggingface():
    try:
        
        client = InferenceClient(
            model="meta-llama/Llama-3.2-1B-Instruct",
            token=hf_token
        )
        
        # 3. 'chat_completion' is the ONLY way to talk to modern providers
        response = client.chat_completion(
            messages=[{"role": "user", "content": "Hi, are you working?"}],
            max_tokens=20
        )
        
        result = response.choices[0].message.content
        
        print("=" * 40)
        print("     HUGGINGFACE CONNECTION")
        print("=" * 40)
        print(f"Status   : Connected!")
        print(f"Model    : Llama-3.2-1B")
        print(f"Response : {result}")
        print("=" * 40)
        return True
        
    except Exception as e:
        # Agar permissions ka error aaye to wo yahan dikhay ga
        print(f"❌ Error Detail: {str(e)}")
        return False

test_huggingface()
```

```python
# ============================================
# STEP 01: Pinecone Connection Test
# ============================================

from pinecone import Pinecone, ServerlessSpec

def test_pinecone_connection():
    try:
        # Pinecone initialize karein (Key .env se pehle hi load ho chuki hai)
        pc = Pinecone(api_key=pinecone_key)
        
        # Maujooda indexes ki list check karein
        active_indexes = pc.list_indexes()
        index_names = [idx.name for idx in active_indexes]
        
        print("=" * 40)
        print("      PINECONE CONNECTION")
        print("=" * 40)
        print(f"Status   : Connected!")
        print(f"Indexes  : {index_names if index_names else 'No indexes found (Normal)'}")
        print("=" * 40)
        return True
    except Exception as e:
        print(f"❌ Pinecone Connection Error: {str(e)}")
        return False

# Function run karein
test_pinecone_connection()
```
